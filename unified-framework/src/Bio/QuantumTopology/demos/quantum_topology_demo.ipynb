## 1. Basic Helical Coordinate Generation

Transform biological sequences to quantum-helical coordinates using the Z Framework's φ-geodesic mapping.

```python
# Create sample DNA sequence
dna_sequence = Seq("ATGCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATC")
print(f"Sample DNA sequence: {dna_sequence}")
print(f"Length: {len(dna_sequence)} bases")

# Generate helical coordinates using optimal curvature k=0.3
coords = generate_helical_coordinates(dna_sequence, k=0.3, hypothetical=True)

print("\n=== Helical Coordinate Results ===")
print(f"Generated {len(coords['x'])} coordinate points")
print(f"X coordinate range: [{coords['x'].min():.3f}, {coords['x'].max():.3f}]")
print(f"Y coordinate range: [{coords['y'].min():.3f}, {coords['y'].max():.3f}]")
print(f"Z coordinate range: [{coords['z'].min():.3f}, {coords['z'].max():.3f}]")
print(f"Geodesic θ range: [{coords['theta'].min():.3f}, {coords['theta'].max():.3f}]")

# Display metadata
metadata = coords['metadata']
print("\n=== Transformation Metadata ===")
print(f"Curvature parameter (k): {metadata['curvature_k']}")
print(f"Golden ratio (φ): {metadata['phi_constant']:.6f}")
print(f"Normalization: {metadata['normalization']}")
print(f"Base mapping: {metadata['base_mapping']}")
```

## 2. Quantum Correlation Analysis

Compute quantum-inspired correlations to identify potential "entanglement" patterns in biological sequences.

```python
# Compute quantum correlations with sliding window
corr_data = compute_quantum_correlations(dna_sequence, window_size=10, k=0.3, hypothetical=True)

print("=== Quantum Correlation Analysis ===")
print(f"Window size: {corr_data['window_size']}")
print(f"Number of correlation windows: {len(corr_data['correlations'])}")
print(f"Mean correlation: {corr_data['mean_correlation']:.4f}")
print(f"Correlation std dev: {corr_data['std_correlation']:.4f}")
print(f"Entanglement threshold: {corr_data['correlation_threshold']:.4f}")
print(f"Number of entangled regions: {np.sum(corr_data['entangled_regions'])}")

# Plot correlation pattern
plt.figure(figsize=(12, 6))
positions = np.arange(1, len(corr_data['correlations']) + 1)
plt.plot(positions, corr_data['correlations'], 'b-', linewidth=2, label='Quantum Correlations')
plt.axhline(y=corr_data['correlation_threshold'], color='red', linestyle='--', 
           label=f'Entanglement Threshold: {corr_data["correlation_threshold"]:.3f}')

# Highlight entangled regions
if np.any(corr_data['entangled_regions']):
    entangled_pos = positions[corr_data['entangled_regions']]
    entangled_corr = corr_data['correlations'][corr_data['entangled_regions']]
    plt.scatter(entangled_pos, entangled_corr, color='red', s=50, 
               label='Potential Entanglement', zorder=5)

plt.xlabel('Sequence Position')
plt.ylabel('Correlation Coefficient')
plt.title('Quantum Correlation Pattern Analysis')
plt.legend()
plt.grid(True, alpha=0.3)
plt.figtext(0.5, 0.02, '⚠️ Experimental/Hypothetical Quantum Analysis', 
           ha='center', color='red', fontsize=10)
plt.tight_layout()
plt.show()
```

## 3. Quantum-Inspired Sequence Alignment

Compare sequences using Bell-type violation metrics instead of traditional alignment methods.

```python
# Create two sequences for comparison
seq1 = Seq("ATGCGATCGATCGATC")
seq2 = Seq("ATGCGATCGATCGATC")  # Identical
seq3 = Seq("CGTAGCTACGTAGCTA")  # Different

print("=== Quantum Alignment Analysis ===")
print(f"Sequence 1: {seq1}")
print(f"Sequence 2: {seq2} (identical)")
print(f"Sequence 3: {seq3} (different)")

# Test different alignment methods
methods = ['bell_violation', 'correlation', 'hybrid']
results = {}

for method in methods:
    print(f"\n--- {method.upper()} METHOD ---")
    
    # Compare identical sequences
    result_identical = quantum_alignment(seq1, seq2, method=method, hypothetical=True)
    results[f'{method}_identical'] = result_identical
    print(f"Seq1 vs Seq2 (identical): {result_identical['alignment_score']:.4f}")
    
    # Compare different sequences
    result_different = quantum_alignment(seq1, seq3, method=method, hypothetical=True)
    results[f'{method}_different'] = result_different
    print(f"Seq1 vs Seq3 (different): {result_different['alignment_score']:.4f}")
    
    # Show method-specific details
    if method == 'bell_violation':
        print(f"  Bell violation strength: {result_different.get('bell_violation', 'N/A')}")
        print(f"  Statistical p-value: {result_different.get('p_value', 'N/A')}")
    elif method == 'correlation':
        print(f"  Correlation similarity: {result_different.get('correlation_similarity', 'N/A')}")

# Create comparison visualization
plt.figure(figsize=(12, 8))

# Plot alignment scores
method_names = ['Bell Violation', 'Correlation', 'Hybrid']
identical_scores = [results[f'{m}_identical']['alignment_score'] for m in methods]
different_scores = [results[f'{m}_different']['alignment_score'] for m in methods]

x = np.arange(len(method_names))
width = 0.35

plt.bar(x - width/2, identical_scores, width, label='Identical Sequences', color='blue', alpha=0.7)
plt.bar(x + width/2, different_scores, width, label='Different Sequences', color='red', alpha=0.7)

plt.xlabel('Alignment Method')
plt.ylabel('Quantum Alignment Score')
plt.title('Quantum Alignment Method Comparison')
plt.xticks(x, method_names)
plt.legend()
plt.grid(True, alpha=0.3)
plt.figtext(0.5, 0.02, '⚠️ Experimental/Hypothetical Quantum Alignment', 
           ha='center', color='red', fontsize=10)
plt.tight_layout()
plt.show()
```

## 4. Bell Violation Analysis

Compute Bell-type violations to detect quantum-like correlations between sequence coordinates.

```python
# Generate coordinates for Bell violation analysis
coords1 = generate_helical_coordinates(seq1, k=0.3, hypothetical=False)
coords3 = generate_helical_coordinates(seq3, k=0.3, hypothetical=False)

# Compute Bell violations
violation_identical, p_identical = compute_bell_violation(coords1, coords1)
violation_different, p_different = compute_bell_violation(coords1, coords3)

print("=== Bell Violation Analysis ===")
print(f"Self-comparison (seq1 vs seq1):")
print(f"  Bell violation: {violation_identical:.4f}")
print(f"  p-value: {p_identical:.4f}")
print(f"  Interpretation: {'Quantum-like' if violation_identical > 0.1 else 'Classical'}")

print(f"\nCross-comparison (seq1 vs seq3):")
print(f"  Bell violation: {violation_different:.4f}")
print(f"  p-value: {p_different:.4f}")
print(f"  Interpretation: {'Quantum-like' if violation_different > 0.1 else 'Classical'}")

print("\nNote: Bell violations > 0 indicate correlations that exceed classical bounds,")
print("suggesting potential quantum-like relationships between sequence positions.")

# Visualize coordinate correlations
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

# Sequence 1 X-Y correlation
ax1.scatter(coords1['x'], coords1['y'], c='blue', alpha=0.6, s=30)
ax1.set_xlabel('X coordinate')
ax1.set_ylabel('Y coordinate')
ax1.set_title('Sequence 1: X-Y Correlation')
ax1.grid(True, alpha=0.3)

# Sequence 3 X-Y correlation
ax2.scatter(coords3['x'], coords3['y'], c='red', alpha=0.6, s=30)
ax2.set_xlabel('X coordinate')
ax2.set_ylabel('Y coordinate')
ax2.set_title('Sequence 3: X-Y Correlation')
ax2.grid(True, alpha=0.3)

# Cross-correlation comparison
ax3.scatter(coords1['x'], coords3['x'], c='purple', alpha=0.6, s=30)
ax3.set_xlabel('Sequence 1: X coordinate')
ax3.set_ylabel('Sequence 3: X coordinate')
ax3.set_title('Cross-sequence X Correlation')
ax3.grid(True, alpha=0.3)

# Bell violation comparison
violations = [violation_identical, violation_different]
p_values = [p_identical, p_different]
labels = ['Self-comparison', 'Cross-comparison']
colors = ['blue', 'red']

bars = ax4.bar(labels, violations, color=colors, alpha=0.7)
ax4.set_ylabel('Bell Violation Strength')
ax4.set_title('Bell Violation Comparison')
ax4.axhline(y=0, color='black', linestyle='-', alpha=0.5)
ax4.grid(True, alpha=0.3)

# Add p-values as text
for i, (bar, p_val) in enumerate(zip(bars, p_values)):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height + 0.001,
             f'p={p_val:.3f}', ha='center', va='bottom', fontsize=10)

plt.figtext(0.5, 0.02, '⚠️ Experimental/Hypothetical Bell Violation Analysis', 
           ha='center', color='red', fontsize=10)
plt.tight_layout()
plt.show()
```

## 5. Low-Complexity Region Analysis

Identify regions of low topological complexity using the κ(n) complexity metric.

```python
# Analyze low-complexity regions
complexity_result = filter_low_complexity_positions(dna_sequence, percentile=20)

print("=== Low-Complexity Region Analysis ===")
print(f"Sequence length: {len(dna_sequence)}")
print(f"Complexity threshold (20th percentile): {complexity_result['threshold']:.4f}")
print(f"Number of low-complexity positions: {len(complexity_result['positions'])}")
print(f"Low-complexity positions: {complexity_result['positions'][:10]}...")  # Show first 10

# Plot complexity profile
plt.figure(figsize=(15, 8))

# Main complexity plot
plt.subplot(2, 1, 1)
positions = np.arange(1, len(dna_sequence) + 1)
kappa_values = complexity_result['kappa_values']

plt.plot(positions, kappa_values, 'b-', linewidth=2, label='κ(n) Complexity')
plt.axhline(y=complexity_result['threshold'], color='red', linestyle='--', 
           label=f'Low-complexity threshold: {complexity_result["threshold"]:.3f}')

# Highlight low-complexity positions
low_complex_mask = kappa_values <= complexity_result['threshold']
plt.scatter(positions[low_complex_mask], kappa_values[low_complex_mask], 
           color='red', s=30, alpha=0.7, label='Low-complexity regions')

plt.xlabel('Sequence Position')
plt.ylabel('κ(n) Complexity Metric')
plt.title('Topological Complexity Profile')
plt.legend()
plt.grid(True, alpha=0.3)

# Sequence visualization with complexity coloring
plt.subplot(2, 1, 2)
base_colors = {'A': 0, 'T': 1, 'G': 2, 'C': 3}
base_numeric = [base_colors.get(str(base), 0) for base in dna_sequence]

# Create color map based on complexity
colors = ['red' if kappa_values[i] <= complexity_result['threshold'] else 'lightblue' 
         for i in range(len(dna_sequence))]

plt.bar(positions, [1]*len(positions), color=colors, alpha=0.7, width=0.8)
plt.xlabel('Sequence Position')
plt.ylabel('Base')
plt.title('Sequence with Low-Complexity Highlighting (Red = Low Complexity)')

# Add base labels
for i, base in enumerate(dna_sequence):
    plt.text(i+1, 0.5, base, ha='center', va='center', fontsize=8, fontweight='bold')

plt.ylim(0, 1)
plt.grid(True, alpha=0.3)

plt.figtext(0.5, 0.02, '⚠️ Experimental Topological Complexity Analysis', 
           ha='center', color='red', fontsize=10)
plt.tight_layout()
plt.show()
```

## 6. Quantum Motif Detection

Detect potential quantum motifs using coherence patterns in helical coordinates.

```python
# Detect quantum motifs
motif_result = detect_quantum_motifs(dna_sequence, motif_length=6, k=0.3, threshold=0.5)

print("=== Quantum Motif Detection ===")
print(f"Motif length: {motif_result['motif_length']}")
print(f"Coherence threshold: {motif_result['threshold']}")
print(f"Number of potential quantum motifs found: {len(motif_result['motifs'])}")

# Display top motifs
if motif_result['motifs']:
    print("\nTop quantum motifs (sorted by coherence):")
    for i, motif in enumerate(motif_result['motifs'][:5]):  # Show top 5
        print(f"{i+1}. Position {motif['position']:2d}: {motif['motif']} "
              f"(coherence: {motif['coherence']:.3f})")
else:
    print("No quantum motifs found above threshold.")

# Visualize motif coherence distribution
if motif_result['motifs']:
    plt.figure(figsize=(15, 10))
    
    # Coherence distribution
    plt.subplot(2, 1, 1)
    coherences = [motif['coherence'] for motif in motif_result['motifs']]
    positions_motif = [motif['position'] for motif in motif_result['motifs']]
    
    plt.bar(range(len(coherences)), coherences, color='purple', alpha=0.7)
    plt.axhline(y=motif_result['threshold'], color='red', linestyle='--', 
               label=f'Threshold: {motif_result["threshold"]}')
    plt.xlabel('Motif Rank')
    plt.ylabel('Quantum Coherence')
    plt.title('Quantum Motif Coherence Distribution')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Motif positions along sequence
    plt.subplot(2, 1, 2)
    all_positions = np.arange(1, len(dna_sequence) + 1)
    motif_mask = np.zeros(len(dna_sequence), dtype=bool)
    
    for motif in motif_result['motifs']:
        start = motif['position']
        end = start + motif['length']
        motif_mask[start:end] = True
    
    # Color sequence positions based on motif presence
    colors = ['purple' if motif_mask[i] else 'lightblue' for i in range(len(dna_sequence))]
    plt.bar(all_positions, [1]*len(all_positions), color=colors, alpha=0.7, width=0.8)
    
    # Add base labels
    for i, base in enumerate(dna_sequence):
        plt.text(i+1, 0.5, base, ha='center', va='center', fontsize=8, fontweight='bold')
    
    plt.xlabel('Sequence Position')
    plt.ylabel('Base')
    plt.title('Sequence with Quantum Motifs Highlighted (Purple = Motif Region)')
    plt.ylim(0, 1)
    plt.grid(True, alpha=0.3)
    
    plt.figtext(0.5, 0.02, '⚠️ Experimental/Hypothetical Quantum Motif Detection', 
               ha='center', color='red', fontsize=10)
    plt.tight_layout()
    plt.show()
else:
    print("\nNo motifs to visualize. Try lowering the threshold or using a longer sequence.")
```

## 7. 3D Helical Visualization

Create 3D visualization of the quantum-helical coordinate transformation.

```python
# Create 3D helical visualization
from mpl_toolkits.mplot3d import Axes3D

# Use a shorter sequence for clearer visualization
viz_seq = Seq("ATGCGATCGATCGATC")
viz_coords = generate_helical_coordinates(viz_seq, k=0.3, hypothetical=False)

# Create 3D plot
fig = plt.figure(figsize=(15, 12))

# Main 3D helix
ax1 = fig.add_subplot(2, 2, 1, projection='3d')
x, y, z = viz_coords['x'], viz_coords['y'], viz_coords['z']

# Plot helical structure
ax1.plot(x, y, z, 'b-', alpha=0.6, linewidth=2, label='Helical backbone')

# Color-code bases
base_colors = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'orange'}
for i, base in enumerate(viz_seq):
    color = base_colors.get(str(base), 'gray')
    ax1.scatter(x[i], y[i], z[i], c=color, s=100, alpha=0.8, label=f'Base {base}' if i == 0 or str(viz_seq[i-1]) != str(base) else '')

ax1.set_xlabel('X (φ-scaled)')
ax1.set_ylabel('Y (φ-scaled)')
ax1.set_zlabel('Z (helical progression)')
ax1.set_title('3D Quantum-Helical Structure')

# X-Y projection
ax2 = fig.add_subplot(2, 2, 2)
for i, base in enumerate(viz_seq):
    color = base_colors.get(str(base), 'gray')
    ax2.scatter(x[i], y[i], c=color, s=100, alpha=0.8)
    ax2.annotate(f'{i+1}', (x[i], y[i]), xytext=(5, 5), textcoords='offset points', fontsize=8)

ax2.plot(x, y, 'b-', alpha=0.5, linewidth=1)
ax2.set_xlabel('X coordinate')
ax2.set_ylabel('Y coordinate')
ax2.set_title('X-Y Projection (Cross-section)')
ax2.grid(True, alpha=0.3)
ax2.axis('equal')

# Z progression
ax3 = fig.add_subplot(2, 2, 3)
positions = np.arange(1, len(viz_seq) + 1)
ax3.plot(positions, z, 'g-', linewidth=2, marker='o', markersize=6, label='Z progression')
ax3.set_xlabel('Sequence Position')
ax3.set_ylabel('Z coordinate')
ax3.set_title('Helical Progression Along Z-axis')
ax3.grid(True, alpha=0.3)
ax3.legend()

# Geodesic angles
ax4 = fig.add_subplot(2, 2, 4)
theta = viz_coords['theta']
ax4.plot(positions, theta, 'm-', linewidth=2, marker='s', markersize=6, label='θ\' geodesic')
ax4.axhline(y=viz_coords['metadata']['phi_constant'], color='red', linestyle='--', 
           label=f'φ = {viz_coords["metadata"]["phi_constant"]:.3f}')
ax4.set_xlabel('Sequence Position')
ax4.set_ylabel('Geodesic Angle θ\'')
ax4.set_title('φ-Geodesic Transformation')
ax4.grid(True, alpha=0.3)
ax4.legend()

plt.figtext(0.5, 0.02, '⚠️ Experimental/Hypothetical Quantum-Helical Visualization', 
           ha='center', color='red', fontsize=10)
plt.tight_layout()
plt.show()

# Print sequence with position mapping
print("\n=== Sequence-to-Coordinate Mapping ===")
print(f"Sequence: {viz_seq}")
print("Position:  ", end="")
for i in range(len(viz_seq)):
    print(f"{i+1:>2}", end="")
print()
print("Base:      ", end="")
for base in viz_seq:
    print(f"{base:>2}", end="")
print()
print("X coord:   ", end="")
for i in range(len(viz_seq)):
    print(f"{x[i]:>5.2f}", end=" ")
print()
print("Y coord:   ", end="")
for i in range(len(viz_seq)):
    print(f"{y[i]:>5.2f}", end=" ")
print()
print("Z coord:   ", end="")
for i in range(len(viz_seq)):
    print(f"{z[i]:>5.2f}", end=" ")
print()
```
