```python
import sympy
import numpy as np
from scipy.stats import chisquare, pearsonr
import mpmath as mp
mp.mp.dps = 50

def adaptive_grid_map(number_str, grid_size):
    mid = len(number_str) // 2
    left = number_str[:mid] or '0'
    right = number_str[mid:] or '0'
    x = int(left) % grid_size
    y = int(right) % grid_size
    return x, y

bits = 256  # Smaller for speed; ~77 digits, grid~38
grid_size = max(8, bits // 3)  # Adjust

moduli = []
for _ in range(500):
    p = sympy.randprime(2**(bits-1), 2**bits)
    q = sympy.randprime(2**(bits-1), 2**bits)
    moduli.append(str(p * q))

coords = [adaptive_grid_map(m, grid_size) for m in moduli]
x_vals, y_vals = zip(*coords)

# Stats
num_bins = min(10, grid_size)
bins = np.linspace(0, grid_size, num_bins + 1)
obs_x = np.histogram(x_vals, bins=bins)[0]
exp = len(x_vals) / num_bins
chi2_x, p_x = chisquare(obs_x, [exp]*num_bins) if np.sum(obs_x) else (0, 1)

# Similar for y; unique, r, etc.
```
