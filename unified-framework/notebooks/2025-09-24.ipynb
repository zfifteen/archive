```python
import mpmath as mp
import numpy as np

mp.mp.dps = 30  # Reduced for speed; sufficient for geodesic/curvature

# Faster divisor count: Trial division with even/odd optimization
def fast_d(n):
    if n < 2:
        return 1
    count = 1
    if n % 2 == 0:
        count += 1
        while n % 2 == 0:
            n //= 2
    i = 3
    sqrt_n = mp.sqrt(n)
    while i <= sqrt_n:
        if n % i == 0:
            count += 1
            while n % i == 0:
                n //= i
        i += 2
    if n > 1:
        count += 1
    return count

def kappa(n):
    d = fast_d(int(n))
    return d * mp.log(n + 1) / mp.exp(2)**2

def theta_prime(n, k_star=0.04449):
    phi = (1 + mp.sqrt(5)) / 2
    return phi * ((n % phi / phi) ** k_star)

# Reduced range: logspace(3,8,num=4) -> k=[1000, 21544, 464159, 10000000]
k_values = np.logspace(3, 8, num=4, dtype=int)

# Known p_k from project docs (pasted-text.txt) for validation
known_p = {
    1000: 7919,
    21544: None,  # Approximate or skip true p if not needed
    464159: None,
    10000000: 179424673
}

results = []
for k in k_values:
    p_true = known_p.get(int(k), None)
    curv = kappa(int(k))
    theta = theta_prime(int(k))
    results.append({'k': int(k), 'p_true': p_true, 'curvature': float(curv), 'geodesic': float(theta)})

print(results)
```
