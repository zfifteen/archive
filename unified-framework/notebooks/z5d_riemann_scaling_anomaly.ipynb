# Z5D Riemann Scaling Anomaly Analysis

This notebook demonstrates the O(1/log²k) correction analysis for Z5D prime predictions.

```python
import sys
sys.path.append('..')

from src.analysis.io_utils import load_white_paper
from src.analysis.z5d_riemann_scaling_anomaly import (
    compute_correction_features,
    fit_correction,
    bootstrap_analysis
)
import matplotlib.pyplot as plt
import numpy as np
```

## Load and examine the data

```python
# Load the white paper data
df = load_white_paper('../docs/WHITE_PAPER_RIEMANN_Z5D_SYMMETRY_FALSIFICATION.md')
print(f"Loaded {len(df)} samples")
print(f"k range: [{df['k'].min()}, {df['k'].max()}]")
df.head()
```

## Compute correction features

```python
df = compute_correction_features(df)

# Plot the 1/log²k feature
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.plot(df['k'], df['ilog2'], 'o-')
plt.xlabel('k (prime index)')
plt.ylabel('1/log²k')
plt.title('Correction Feature vs k')
plt.yscale('log')

plt.subplot(1, 2, 2)
plt.plot(df['logk'], df['err_rel'], 'o-')
plt.xlabel('log(k)')
plt.ylabel('Relative Error')
plt.title('Relative Error vs log(k)')

plt.tight_layout()
plt.show()
```

## Fit correction model

```python
# Fit multiplicative correction
result = fit_correction(df, form="multiplicative", holdout_pct=20)

print(f"Estimated β: {result['beta']:.6f}")
print(f"RMSE improvement: {result['metric_improvement']['rmse_delta']:.6f}")
print(f"Sample sizes: {result['n']} test, {result['n_train']} train")
```

## Bootstrap analysis

```python
# Run bootstrap analysis (reduced samples for notebook speed)
bootstrap_results = bootstrap_analysis(df, n_bootstrap=500, holdout_pct=20)

print(f"Bootstrap β mean: {bootstrap_results['bootstrap']['beta_mean']:.6f}")
print(f"Bootstrap β 95% CI: [{bootstrap_results['bootstrap']['beta_ci'][0]:.6f}, {bootstrap_results['bootstrap']['beta_ci'][1]:.6f}]")
print(f"p-value: {bootstrap_results['bootstrap']['p_value']:.6f}")
print(f"Significant: {bootstrap_results['bootstrap']['significance']}")
```

## Visualization

```python
# Extract bootstrap results for plotting
betas = [r['beta'] for r in bootstrap_results['raw_results']]
rmse_deltas = [r['metric_improvement']['rmse_delta'] for r in bootstrap_results['raw_results']]

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(betas, bins=20, alpha=0.7, edgecolor='black')
plt.xlabel('β coefficient')
plt.ylabel('Frequency')
plt.title('Bootstrap Distribution of β')
plt.axvline(bootstrap_results['bootstrap']['beta_mean'], color='red', linestyle='--', label='Mean')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(rmse_deltas, bins=20, alpha=0.7, edgecolor='black')
plt.xlabel('RMSE Improvement')
plt.ylabel('Frequency')
plt.title('Bootstrap Distribution of RMSE Improvement')
plt.axvline(0, color='red', linestyle='-', label='No improvement')
plt.axvline(bootstrap_results['bootstrap']['rmse_delta_mean'], color='orange', linestyle='--', label='Mean')
plt.legend()

plt.tight_layout()
plt.show()
```
