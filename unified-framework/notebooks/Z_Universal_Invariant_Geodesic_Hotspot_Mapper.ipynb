# Z Universal Invariant Geodesic Hotspot Mapper

## Interactive Gene/Chromosome Analysis Tool

This notebook provides a proof-of-concept (PoC) tool leveraging the Z Biopython module to perform real-time, interactive gene- or chromosome-scale analysis using the universal invariant framework.

### Key Features:
- **Load and process full genes or small chromosomes from FASTA**
- **Compute Z-invariant/geodesic coordinates and identify prime density geodesic hotspots**
- **Overlay and correlate with biological annotations (GFF/BED)**
- **Interactive interface with instant feedback and statistics**
- **Linear time performance demonstration**

### Mathematical Foundation:
- **Universal equation**: `Z = A(B/c)` adapted as `Z = n(Δₙ / Δₘₐₓ)`
- **Complexity metric**: `κ(n) = d(n) · ln(n+1) / e²`
- **Geodesic mapping**: `θ'(n, k) = φ·{n/φ}^k` with k* ≈ 0.3
- **Target enhancement**: conditional prime density improvement under canonical benchmark methodology improvement

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import ipywidgets as widgets
from IPython.display import display, HTML
import io
import base64
from pathlib import Path
import time
import warnings

# Import Z Framework components
import sys
sys.path.append('../..')

from src.Bio.QuantumTopology import ZGeodesicHotspotMapper
from Bio.Seq import Seq
from Bio import SeqIO

# Configure display
plt.style.use('default')
warnings.filterwarnings('ignore', category=UserWarning)

print("✅ Z Universal Invariant Geodesic Hotspot Mapper loaded successfully!")
print("📊 Ready for interactive gene/chromosome analysis")
```

## 1. Initialize the Z Geodesic Hotspot Mapper

Configure the mapper with optimal parameters for φ-geodesic transformation and prime density enhancement.

```python
# Initialize the Z Geodesic Hotspot Mapper
mapper = ZGeodesicHotspotMapper(
    k_optimal=0.3,  # Optimal curvature for 15% density enhancement
    modulus=1.618   # Golden ratio φ
)

print(f"🔧 Mapper initialized with:")
print(f"   • Optimal curvature k* = {mapper.k_optimal}")
print(f"   • Golden ratio φ = {mapper.phi:.6f}")
print(f"   • Normalization constant e² = {mapper.e_squared:.6f}")
print(f"   • Target density enhancement: 15%")
```

## 2. Interactive File Loading and Sequence Selection

```python
# Create sample sequences for demonstration
sample_sequences = {
    'Sample Gene (100bp)': Seq('ATGAAAGCGTTGCTGAAGCATGTAGCGGATCCGTTAGCTAGCGATCGCTAG' + 
                              'CTAGCGATCGTAGCTAGCGATCGTAGCGATCGTAGCGATCGTAGCGATCGAT'),
    'Sample Gene (200bp)': Seq('ATGAAAGCGTTGCTGAAGCATGTAGCGGATCCGTTAGCTAGCGATCGCTAG' * 4),
    'Random Test (500bp)': Seq(''.join(np.random.choice(['A', 'T', 'G', 'C'], 500))),
    'Custom Input': None  # Will be filled by user input
}

# File upload widget
file_upload = widgets.FileUpload(
    accept='.fasta,.fa,.fas',
    multiple=False,
    description='Upload FASTA'
)

# Sequence selection dropdown
sequence_dropdown = widgets.Dropdown(
    options=list(sample_sequences.keys()),
    value='Sample Gene (100bp)',
    description='Select Sequence:'
)

# Custom sequence input
custom_sequence_input = widgets.Textarea(
    value='ATGCGATCGATCGTAGCGATC',
    placeholder='Enter custom DNA sequence...',
    description='Custom DNA:',
    disabled=True
)

# Enable/disable custom input based on selection
def on_sequence_change(change):
    custom_sequence_input.disabled = (change['new'] != 'Custom Input')
    if change['new'] == 'Custom Input':
        sample_sequences['Custom Input'] = Seq(custom_sequence_input.value)

sequence_dropdown.observe(on_sequence_change, names='value')
custom_sequence_input.observe(lambda change: setattr(sample_sequences, 'Custom Input', Seq(change['new'])), names='value')

print("📁 File Upload and Sequence Selection:")
display(widgets.VBox([
    widgets.HTML("<b>Option 1: Upload FASTA file</b>"),
    file_upload,
    widgets.HTML("<b>Option 2: Select sample sequence</b>"),
    sequence_dropdown,
    custom_sequence_input
]))
```

## 3. Interactive Parameter Controls

Adjust geodesic parameters and analysis settings in real-time.

```python
# Interactive parameter controls
k_slider = widgets.FloatSlider(
    value=0.3,
    min=0.01,
    max=1.0,
    step=0.01,
    description='Curvature k:',
    continuous_update=False,
    readout_format='.3f'
)

modulus_dropdown = widgets.Dropdown(
    options=[('Golden Ratio φ', 1.618), ('Euler e', 2.718), ('Pi π', 3.14159), ('Custom', 0)],
    value=1.618,
    description='Modulus:'
)

custom_modulus = widgets.FloatText(
    value=1.618,
    description='Custom Mod:',
    disabled=True
)

density_threshold = widgets.FloatSlider(
    value=1.5,
    min=1.0,
    max=3.0,
    step=0.1,
    description='Density Threshold:',
    continuous_update=False,
    readout_format='.1f'
)

window_size = widgets.IntSlider(
    value=100,
    min=20,
    max=500,
    step=10,
    description='Window Size:',
    continuous_update=False
)

# Enable custom modulus when selected
def on_modulus_change(change):
    custom_modulus.disabled = (change['new'] != 0)
    if change['new'] == 0:
        mapper.modulus = custom_modulus.value
    else:
        mapper.modulus = change['new']

modulus_dropdown.observe(on_modulus_change, names='value')

print("🎛️ Interactive Parameter Controls:")
display(widgets.VBox([
    widgets.HTML("<b>Geodesic Transformation Parameters</b>"),
    widgets.HBox([k_slider, modulus_dropdown, custom_modulus]),
    widgets.HTML("<b>Hotspot Detection Parameters</b>"),
    widgets.HBox([density_threshold, window_size])
]))
```

## 4. Real-Time Analysis and Visualization

Perform Z-invariant coordinate computation and hotspot detection with live updates.

```python
# Analysis results storage
analysis_results = {
    'coordinates': None,
    'hotspots': None,
    'performance': None
}

# Output area for results
output_area = widgets.Output()

def run_analysis(change=None):
    """Run the complete Z-invariant geodesic analysis."""
    
    with output_area:
        output_area.clear_output(wait=True)
        
        # Get current sequence
        if sequence_dropdown.value == 'Custom Input':
            current_sequence = Seq(custom_sequence_input.value)
        else:
            current_sequence = sample_sequences[sequence_dropdown.value]
        
        if not current_sequence or len(current_sequence) == 0:
            print("❌ Please select or enter a valid sequence")
            return
        
        print(f"🔄 Analyzing sequence: {len(current_sequence)} bases")
        print(f"📊 Parameters: k={k_slider.value:.3f}, modulus={mapper.modulus:.3f}")
        
        # Update mapper parameters
        mapper.k_optimal = k_slider.value
        
        start_time = time.time()
        
        try:
            # Compute Z-invariant coordinates
            print("🧮 Computing Z-invariant geodesic coordinates...")
            coordinates = mapper.compute_z_invariant_coordinates(current_sequence, k=k_slider.value)
            
            # Detect prime hotspots
            print("🎯 Detecting prime density hotspots...")
            hotspots = mapper.detect_prime_hotspots(
                coordinates, 
                density_threshold=density_threshold.value,
                window_size=window_size.value
            )
            
            total_time = time.time() - start_time
            
            # Store results
            analysis_results['coordinates'] = coordinates
            analysis_results['hotspots'] = hotspots
            analysis_results['performance'] = {
                'total_time': total_time,
                'time_per_base': total_time / len(current_sequence),
                'sequence_length': len(current_sequence)
            }
            
            # Display results summary
            print(f"\n✅ Analysis completed in {total_time:.4f} seconds")
            print(f"⚡ Performance: {total_time/len(current_sequence)*1000:.3f} ms/base")
            print(f"🎯 Found {hotspots['total_hotspots']} prime density hotspots")
            print(f"🧬 Sequence complexity range: {coordinates['complexity'].min():.3f} - {coordinates['complexity'].max():.3f}")
            
            if hotspots['total_hotspots'] > 0:
                avg_density = np.mean([h['enhanced_density'] for h in hotspots['hotspots']])
                print(f"📈 Average hotspot density enhancement: {avg_density:.2f}x")
            
            # Create visualization
            create_visualization()
            
        except Exception as e:
            print(f"❌ Analysis failed: {str(e)}")
            import traceback
            traceback.print_exc()

# Run button
run_button = widgets.Button(
    description='🚀 Run Analysis',
    button_style='primary',
    tooltip='Execute Z-invariant geodesic hotspot analysis'
)
run_button.on_click(run_analysis)

print("🚀 Ready for Analysis:")
display(widgets.VBox([run_button, output_area]))
```

## 5. Interactive Visualization Dashboard

```python
def create_visualization():
    """Create interactive visualization of analysis results."""
    
    if not analysis_results['coordinates'] or not analysis_results['hotspots']:
        print("❌ No analysis results to visualize. Please run analysis first.")
        return
    
    coords = analysis_results['coordinates']
    hotspots = analysis_results['hotspots']
    
    # Create subplots
    fig = make_subplots(
        rows=2, cols=2,
        subplot_titles=(
            'Geodesic Coordinate Mapping', 
            'Z-Invariant Transform',
            'Prime Density Analysis',
            '3D Helical Coordinates'
        ),
        specs=[[{'type': 'scatter'}, {'type': 'scatter'}],
               [{'type': 'scatter'}, {'type': 'scatter3d'}]],
        vertical_spacing=0.12
    )
    
    positions = coords['positions']
    
    # 1. Geodesic coordinate mapping
    fig.add_trace(
        go.Scatter(
            x=positions,
            y=coords['geodesic_coords'],
            mode='lines+markers',
            name='Geodesic θ\'(n,k)',
            line=dict(color='blue', width=2),
            marker=dict(size=4)
        ),
        row=1, col=1
    )
    
    # 2. Z-invariant transform
    fig.add_trace(
        go.Scatter(
            x=positions,
            y=coords['z_invariant'],
            mode='lines+markers',
            name='Z-invariant',
            line=dict(color='red', width=2),
            marker=dict(size=4)
        ),
        row=1, col=2
    )
    
    # Add complexity overlay
    fig.add_trace(
        go.Scatter(
            x=positions,
            y=coords['complexity'],
            mode='lines',
            name='Complexity κ(n)',
            line=dict(color='orange', width=1, dash='dot'),
            yaxis='y2'
        ),
        row=1, col=2
    )
    
    # 3. Prime density analysis
    # Plot prime positions
    prime_positions = hotspots['prime_positions']
    if len(prime_positions) > 0:
        prime_mask = np.isin(positions, prime_positions)
        fig.add_trace(
            go.Scatter(
                x=positions[prime_mask],
                y=coords['geodesic_coords'][prime_mask],
                mode='markers',
                name='Prime Positions',
                marker=dict(size=8, color='green', symbol='diamond')
            ),
            row=2, col=1
        )
    
    # Highlight hotspots
    for i, hotspot in enumerate(hotspots['hotspots']):
        fig.add_vrect(
            x0=hotspot['start'], x1=hotspot['end'],
            fillcolor='rgba(255, 0, 0, 0.2)',
            layer='below',
            line_width=0,
            annotation_text=f"Hotspot {i+1}<br>Density: {hotspot['enhanced_density']:.2f}x",
            annotation_position="top",
            row=2, col=1
        )
    
    # Base geodesic line for prime plot
    fig.add_trace(
        go.Scatter(
            x=positions,
            y=coords['geodesic_coords'],
            mode='lines',
            name='Geodesic Base',
            line=dict(color='lightblue', width=1),
            showlegend=False
        ),
        row=2, col=1
    )
    
    # 4. 3D Helical coordinates
    fig.add_trace(
        go.Scatter3d(
            x=coords['x'],
            y=coords['y'],
            z=coords['z'],
            mode='markers+lines',
            name='Helical Path',
            marker=dict(
                size=4,
                color=coords['complexity'],
                colorscale='Viridis',
                colorbar=dict(title='Complexity κ(n)')
            ),
            line=dict(color='navy', width=3)
        ),
        row=2, col=2
    )
    
    # Update layout
    fig.update_layout(
        title=f'Z Universal Invariant Geodesic Hotspot Analysis<br>' +
              f'Sequence Length: {len(positions)} bp | ' +
              f'Hotspots Found: {hotspots["total_hotspots"]} | ' +
              f'Processing Time: {analysis_results["performance"]["total_time"]:.3f}s',
        height=800,
        showlegend=True,
        legend=dict(x=1.02, y=1)
    )
    
    # Update axes labels
    fig.update_xaxes(title_text="Position (bp)", row=1, col=1)
    fig.update_yaxes(title_text="θ'(n,k)", row=1, col=1)
    
    fig.update_xaxes(title_text="Position (bp)", row=1, col=2)
    fig.update_yaxes(title_text="Z-invariant", row=1, col=2)
    
    fig.update_xaxes(title_text="Position (bp)", row=2, col=1)
    fig.update_yaxes(title_text="Geodesic Coordinate", row=2, col=1)
    
    fig.show()

# Visualization control button
viz_button = widgets.Button(
    description='📊 Update Visualization',
    button_style='info',
    tooltip='Refresh visualization with current results'
)
viz_button.on_click(lambda x: create_visualization())

print("📊 Visualization Controls:")
display(viz_button)
```

## 6. Performance Benchmarking

Demonstrate linear time scaling for laptop-scale analysis.

```python
def run_performance_benchmark():
    """Run performance benchmark across different sequence lengths."""
    
    print("🚀 Running performance benchmark...")
    print("📊 Testing linear scaling across multiple sequence lengths")
    
    # Define test sequence lengths
    test_lengths = [100, 500, 1000, 2500, 5000, 10000]
    
    # Run benchmark
    benchmark_results = mapper.benchmark_performance(test_lengths, num_trials=3)
    
    # Display results
    print("\n📈 Benchmark Results:")
    print(f"{'Length (bp)':<12} {'Time (s)':<12} {'ms/base':<12} {'Linear Fit':<12}")
    print("-" * 50)
    
    for i, length in enumerate(test_lengths):
        time_s = benchmark_results['processing_times'][i]
        ms_per_base = benchmark_results['times_per_base'][i] * 1000
        print(f"{length:<12} {time_s:<12.4f} {ms_per_base:<12.3f}")
    
    if 'linear_fit' in benchmark_results:
        fit = benchmark_results['linear_fit']
        print(f"\n📊 Linear Scaling Analysis:")
        print(f"   • R² = {fit['r_squared']:.4f} (target: >0.95)")
        print(f"   • p-value = {fit['p_value']:.2e}")
        print(f"   • Linear scaling: {'✅ PASSED' if fit['linear_scaling'] else '❌ FAILED'}")
        
        # Estimate performance for larger sequences
        brca1_time = fit['slope'] * 80000 + fit['intercept']  # BRCA1 ~80kb
        chr21_time = fit['slope'] * 46000000 + fit['intercept']  # Chr21 ~46Mb
        
        print(f"\n🧬 Projected Performance:")
        print(f"   • BRCA1 gene (~80kb): {brca1_time:.2f} seconds")
        print(f"   • Chromosome 21 (~46Mb): {chr21_time/60:.1f} minutes")
    
    # Create performance visualization
    fig = go.Figure()
    
    # Plot measured times
    fig.add_trace(go.Scatter(
        x=test_lengths,
        y=benchmark_results['processing_times'],
        mode='markers+lines',
        name='Measured Performance',
        marker=dict(size=10, color='blue')
    ))
    
    # Plot linear fit if available
    if 'linear_fit' in benchmark_results:
        fit_line = [fit['slope'] * x + fit['intercept'] for x in test_lengths]
        fig.add_trace(go.Scatter(
            x=test_lengths,
            y=fit_line,
            mode='lines',
            name=f'Linear Fit (R²={fit["r_squared"]:.3f})',
            line=dict(color='red', dash='dash')
        ))
    
    fig.update_layout(
        title='Z Framework Performance Scaling',
        xaxis_title='Sequence Length (bp)',
        yaxis_title='Processing Time (seconds)',
        xaxis_type='log',
        yaxis_type='log',
        height=500
    )
    
    fig.show()
    
    return benchmark_results

# Benchmark button
benchmark_button = widgets.Button(
    description='⚡ Run Performance Benchmark',
    button_style='warning',
    tooltip='Benchmark linear scaling performance'
)
benchmark_button.on_click(lambda x: run_performance_benchmark())

print("⚡ Performance Benchmarking:")
display(benchmark_button)
```

## 7. Export and Statistical Analysis

```python
def export_analysis_results():
    """Export current analysis results to files."""
    
    if not analysis_results['coordinates'] or not analysis_results['hotspots']:
        print("❌ No analysis results to export. Please run analysis first.")
        return
    
    try:
        # Export to CSV
        timestamp = int(time.time())
        base_path = Path(f"z_geodesic_analysis_{timestamp}")
        
        mapper.export_results(
            analysis_results['hotspots'],
            analysis_results['coordinates'],
            base_path,
            format='csv'
        )
        
        # Also export as JSON
        mapper.export_results(
            analysis_results['hotspots'],
            analysis_results['coordinates'],
            base_path,
            format='json'
        )
        
        print(f"✅ Results exported to:")
        print(f"   • {base_path}.hotspots.csv")
        print(f"   • {base_path}.coordinates.csv")
        print(f"   • {base_path}.json")
        
        # Display statistical summary
        display_statistical_summary()
        
    except Exception as e:
        print(f"❌ Export failed: {str(e)}")

def display_statistical_summary():
    """Display comprehensive statistical summary."""
    
    coords = analysis_results['coordinates']
    hotspots = analysis_results['hotspots']
    perf = analysis_results['performance']
    
    print("\n📊 Statistical Summary:")
    print("=" * 50)
    
    # Sequence statistics
    print(f"🧬 Sequence Analysis:")
    print(f"   • Length: {perf['sequence_length']} bp")
    print(f"   • Processing time: {perf['total_time']:.4f} seconds")
    print(f"   • Throughput: {perf['time_per_base']*1000:.3f} ms/base")
    
    # Z-invariant statistics
    z_stats = {
        'mean': np.mean(coords['z_invariant']),
        'std': np.std(coords['z_invariant']),
        'min': np.min(coords['z_invariant']),
        'max': np.max(coords['z_invariant'])
    }
    
    print(f"\n🔢 Z-Invariant Transform:")
    print(f"   • Mean: {z_stats['mean']:.3f}")
    print(f"   • Std Dev: {z_stats['std']:.3f}")
    print(f"   • Range: [{z_stats['min']:.3f}, {z_stats['max']:.3f}]")
    
    # Geodesic statistics
    geo_stats = {
        'mean': np.mean(coords['geodesic_coords']),
        'std': np.std(coords['geodesic_coords']),
        'variance_ratio': np.std(coords['geodesic_coords']) / 0.118  # Target variance
    }
    
    print(f"\n🌀 Geodesic Coordinates:")
    print(f"   • Mean: {geo_stats['mean']:.3f}")
    print(f"   • Std Dev: {geo_stats['std']:.3f} (target: 0.118)")
    print(f"   • Variance ratio: {geo_stats['variance_ratio']:.3f}")
    
    # Hotspot statistics
    print(f"\n🎯 Prime Hotspot Analysis:")
    print(f"   • Total hotspots: {hotspots['total_hotspots']}")
    print(f"   • Total primes: {hotspots['total_primes']}")
    print(f"   • Prime density: {hotspots['total_primes']/perf['sequence_length']:.4f}")
    
    if hotspots['total_hotspots'] > 0:
        densities = [h['enhanced_density'] for h in hotspots['hotspots']]
        avg_enhancement = np.mean(densities)
        max_enhancement = np.max(densities)
        
        print(f"   • Average enhancement: {avg_enhancement:.2f}x")
        print(f"   • Maximum enhancement: {max_enhancement:.2f}x")
        print(f"   • Enhancement vs target (1.15x): {avg_enhancement/1.15:.2f}")
    
    # Complexity analysis
    complexity_stats = {
        'mean': np.mean(coords['complexity']),
        'entropy': -np.sum(coords['complexity'] * np.log(coords['complexity'] + 1e-10)) / len(coords['complexity'])
    }
    
    print(f"\n🔍 Complexity Analysis:")
    print(f"   • Mean κ(n): {complexity_stats['mean']:.3f}")
    print(f"   • Complexity entropy: {complexity_stats['entropy']:.3f}")

# Export button
export_button = widgets.Button(
    description='💾 Export Results',
    button_style='success',
    tooltip='Export analysis results and display statistics'
)
export_button.on_click(lambda x: export_analysis_results())

print("💾 Export and Statistics:")
display(export_button)
```

## 8. Biological Annotation Integration (Optional)

Load and correlate with GFF/BED annotation files.

```python
# Annotation file upload
annotation_upload = widgets.FileUpload(
    accept='.gff,.gff3,.bed',
    multiple=False,
    description='Upload Annotations'
)

# Sample annotation data (for demonstration)
sample_annotations = pd.DataFrame({
    'seqid': ['seq1'] * 5,
    'source': ['prediction'] * 5,
    'type': ['gene', 'exon', 'CDS', 'promoter', 'enhancer'],
    'start': [10, 25, 30, 5, 80],
    'end': [50, 45, 40, 15, 120],
    'score': [0.9, 0.8, 0.95, 0.7, 0.6],
    'strand': ['+', '+', '+', '+', '+'],
    'phase': [None, None, 0, None, None],
    'attributes': ['ID=gene1', 'ID=exon1', 'ID=cds1', 'ID=prom1', 'ID=enh1']
})

def analyze_annotation_correlation():
    """Analyze correlation between hotspots and annotations."""
    
    if not analysis_results['hotspots']:
        print("❌ No hotspot analysis results. Please run analysis first.")
        return
    
    # Use sample annotations for demonstration
    annotations = sample_annotations.copy()
    
    print("🧬 Analyzing hotspot-annotation correlations...")
    print(f"📝 Using {len(annotations)} sample annotations")
    
    # Perform correlation analysis
    correlation_results = mapper.correlate_hotspots_with_annotations(
        analysis_results['hotspots'],
        annotations,
        overlap_threshold=0.3
    )
    
    print(f"\n📊 Correlation Results:")
    print(f"   • Total correlations: {correlation_results['total_correlations']}")
    print(f"   • Correlation rate: {correlation_results['correlation_rate']:.1%}")
    print(f"   • Annotation types: {', '.join(correlation_results['annotation_types'])}")
    
    if correlation_results['correlations']:
        print(f"\n🎯 Significant Correlations:")
        for i, corr in enumerate(correlation_results['correlations'][:5]):
            print(f"   {i+1}. {corr['annotation_type']} at {corr['annotation_start']}-{corr['annotation_end']}")
            print(f"      Hotspot density: {corr['hotspot_density']:.2f}x, Overlap: {corr['overlap_fraction']:.1%}")
    
    # Create correlation visualization
    create_annotation_plot(annotations, correlation_results)
    
    return correlation_results

def create_annotation_plot(annotations, correlations):
    """Create visualization with annotation overlays."""
    
    if not analysis_results['coordinates']:
        return
    
    coords = analysis_results['coordinates']
    hotspots = analysis_results['hotspots']
    
    fig = go.Figure()
    
    # Plot geodesic coordinates
    fig.add_trace(go.Scatter(
        x=coords['positions'],
        y=coords['geodesic_coords'],
        mode='lines',
        name='Geodesic Coordinates',
        line=dict(color='blue', width=2)
    ))
    
    # Add annotation tracks
    for _, ann in annotations.iterrows():
        fig.add_shape(
            type="rect",
            x0=ann['start'], x1=ann['end'],
            y0=-0.2, y1=-0.1,
            fillcolor="green",
            opacity=0.7,
            layer="below",
            line_width=0,
        )
        
        fig.add_annotation(
            x=(ann['start'] + ann['end']) / 2,
            y=-0.15,
            text=ann['type'],
            showarrow=False,
            font=dict(size=8)
        )
    
    # Add hotspot overlays
    for i, hotspot in enumerate(hotspots['hotspots']):
        fig.add_vrect(
            x0=hotspot['start'], x1=hotspot['end'],
            fillcolor='rgba(255, 0, 0, 0.2)',
            layer='below',
            line_width=0,
            annotation_text=f"H{i+1}",
            annotation_position="top"
        )
    
    fig.update_layout(
        title='Geodesic Hotspots with Biological Annotations',
        xaxis_title='Position (bp)',
        yaxis_title='Geodesic Coordinate',
        height=500,
        yaxis_range=[-0.3, max(coords['geodesic_coords']) * 1.1]
    )
    
    fig.show()

# Annotation analysis button
annotation_button = widgets.Button(
    description='🧬 Analyze Annotations',
    button_style='info',
    tooltip='Correlate hotspots with biological annotations'
)
annotation_button.on_click(lambda x: analyze_annotation_correlation())

print("🧬 Biological Annotation Integration:")
display(widgets.VBox([
    widgets.HTML("<b>Upload annotation file (GFF/BED)</b>"),
    annotation_upload,
    widgets.HTML("<b>Or use sample annotations for demonstration</b>"),
    annotation_button
]))
```
